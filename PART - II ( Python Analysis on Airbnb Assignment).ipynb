### Task 3

Plot a visualization to show the number of listings for each `room_type`

Which `room_type` has the highest number of listings?

```python
# Group the data by room_type and count the number of listings in each category
room_type_counts = df['room_type'].value_counts()

# Create a bar plot to visualize the number of listings for each room_type
plt.figure(figsize=(12, 10))
room_type_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Listings by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Number of Listings')



# Find the room_type with the highest number of listings
highest_count_room_type = room_type_counts.idxmax()
highest_count = room_type_counts.max()

# Annotate the room_type with the highest number of listings
plt.annotate(f'Highest: {highest_count_room_type} ({highest_count} listings)', 
             xy=(room_type_counts.index.get_loc(highest_count_room_type), highest_count), 
             xytext=(10, 20), textcoords='offset points',
             arrowprops=dict(arrowstyle='->', color='red'))

# Show the plot
plt.tight_layout()
plt.show()

# Print the room_type with the highest number of listings
print(f"The room type with the highest number of listings is '{highest_count_room_type}' with {highest_count} listings.")
```

**Observation** -The room type with the highest number of listings is 'Entire home/apt' with 4223 listings.

### Task 4

You want to observe the relationship between the reviews(given by `number_of_reviews`) and the ratings received (`review_scores_rating`) by different AirBnB listings.

For this, plot both a scatterplot and a jointplot. What can you say about the relationship between ratings and reviews?

```python
# Create a scatterplot
sns.scatterplot(data=df, x='number_of_reviews', y='review_scores_rating' , hue = 'room_type')
plt.title('Scatterplot: Number of Reviews vs. Review Scores Rating')
plt.xlabel('Number of Reviews')
plt.ylabel('Review Scores Rating')
plt.legend(title = "Room Type")
plt.show()
```

```python
# Create a jointplot
sns.jointplot(data=df, x='number_of_reviews', y='review_scores_rating' , hue = 'room_type')
plt.suptitle('Jointplot: Number of Reviews vs. Review Scores Rating')
plt.xlabel('Number of Reviews')
plt.ylabel('Review Scores Rating')
plt.legend(title = "Room Type")
plt.show()
```

**Observation** - From the above graph we can say that mostly guest choose private room or entire home/apt on the basis of total number of reviews and  average rating.

### Task 5

AirBnB has been adding quite a few listings in the city of Amseterdam since they started operating in 2008. Plot a lineplot to observe how the average `review_scores_rating` has changed across the different years (use `host_since_Year` column) AirBnB has been operating in the city.

```python
# Extract the 'host_since_Year' and 'review_scores_rating' columns
data = df[['host_since_Year', 'review_scores_rating']]

# Group the data by 'host_since_Year' and calculate the average review_scores_rating for each year
average_ratings_by_year = data.groupby('host_since_Year')['review_scores_rating'].mean().reset_index()

# Create a line plot to show how the average review_scores_rating has changed over the years
sns.lineplot(data=average_ratings_by_year, x='host_since_Year', y='review_scores_rating', marker='o', color='skyblue')
plt.title('Average Review Scores Rating Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Review Scores Rating')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

**Observation** -The  line plot allows you to see the trend in average review scores over the years.There was a great downfall in 2008 and after that it rose in 2010 but for a while but  after that  there was a downfall trend throughout these years in Amseterdam.

### Task 6

You wish to identify the relationship between the various ratings each of the listing has received from the customers.
These ratings have been summarized below:

```python
review_columns = ['review_scores_rating',
 'review_scores_accuracy',
 'review_scores_cleanliness',
 'review_scores_checkin',
 'review_scores_communication',
 'review_scores_location',
 'review_scores_value']
```

Plot a heatmap of the correlation matrix of the above ratings variables and document your observations. You can use this link to understand these variables further - https://www.airbnb.co.in/help/article/1257/star-ratings

```python
# Select the relevant review columns
review_columns = ['review_scores_rating',
                  'review_scores_accuracy',
                  'review_scores_cleanliness',
                  'review_scores_checkin',
                  'review_scores_communication',
                  'review_scores_location',
                  'review_scores_value']

# Create a subset of the DataFrame with only the selected review columns
review_data = df[review_columns]

# Calculate the correlation matrix
correlation_matrix = review_data.corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', linewidths=0.5)
plt.title('Correlation Heatmap of Airbnb Review Ratings')
plt.tight_layout()
plt.show()
```

**Observation** - Strong positive correlations (values close to 1) suggest that when one aspect of the review score is high, others tend to be high as well.Strong negative correlations (values close to -1) indicate that when one aspect of the review score is high, others tend to be low, and vice versa.Weak or near-zero correlations (values close to 0) suggest little to no relationship between two aspects of the review score in Amseterdam.

### Task 7 

Analyze the listings' prices across the following neighborhoods using a categorical boxplot
- 'Westerpark', 'Oud-Noord', 'Noord-West','Zuid'

```python
nc = ['Westerpark', 'Oud-Noord', 'Noord-West','Zuid']
```

```python
# Create a subset of the DataFrame containing only the specified neighborhoods
neighborhood_data = df[df['neighbourhood'].isin(nc)]
```

```python
# Create a categorical boxplot to analyze prices by neighborhood
plt.figure(figsize=(10, 6))
sns.boxplot(data=neighborhood_data, x='neighbourhood', y='price', order=nc)
plt.title('Listings Prices by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Price')
plt.xticks(rotation=45)

# Show the boxplot
plt.tight_layout()
plt.show()
```

**Observation** -  This visualization helps you gain insights into the relative affordability of different neighborhoods. You might notice that some neighborhoods are more budget-friendly, while others cater to higher-end listings.
